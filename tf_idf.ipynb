```python
import pandas as pd


df = pd.read_csv('final_preprocessed_df.csv')

# Display the first 5 rows
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
df_combined = df.groupby('listing_id')['comments'].apply(lambda x: ' '.join(x)).reset_index()
vocab = df_combined['comments']
vectorizer = TfidfVectorizer()
tf_idf_matrix = vectorizer.fit_transform(vocab)
vectorizer.get_feature_names_out()
print(tf_idf_matrix.shape)
tfidf_df = pd.DataFrame(tf_idf_matrix.toarray(), columns=vectorizer.get_feature_names_out(), index=df_combined['listing_id'])
tfidf_df.head()
```
